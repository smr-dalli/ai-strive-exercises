```
# Start writing code here...
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('Aus-temp.csv')
df.head()
```

```
df[['Year','Month','Day']] = df.Date.str.split('-',expand=True)
df
```

```
sns.lineplot(df.Year,df.Temp)
```

```
df['1 day shift'] = df['Temp'].shift(1)
df['2 day shift'] = df['Temp'].shift(2)
df['3 day shift'] = df['Temp'].shift(3)
df
```
