```
# Start writing code here...
import pandas as pd
import numpy as np
shops_df = pd.read_csv('shops.csv')
shops_df.duplicated()
```

```
shops_df.head()
```

```
shops = shops_df.shop_name.str.split(' ',expand = True)
shops.columns = ['x1','city','code','shoping_mall','type','x2','x3','x4']
shops.head()
```

```
shops.drop(['x1','code','shoping_mall','type','x2','x3','x4'],axis = 'columns',inplace = True)
shops.head()
```

```
shops.at[42,'city'] = 'St. Petersburg'
shops.at[43,'city'] = 'St. Petersburg'
shops.at[20,'city'] = 'Moscow'
shops.at[55,'city'] = np.nan
shops.at[12,'city'] = np.nan
shops.at[9,'city'] = np.nan
shops.isna().sum()
```

```
shops.dropna(inplace=True)
shops.isna().sum()
```

```

from geopy.geocoders import Nominatim
```

```
geolocator = Nominatim(user_agent='mohan')
```

```
# shops['location'] = shops['city'].apply(geolocator.geocode) 
# shops
```

```
shops.at[39,'city'] = 'Rostov-on-Don'
shops.at[40,'city'] = 'Rostov-on-Don'
shops.at[41,'city'] = 'Rostov-on-Don'
```

```
shops['location'] = shops['city'].apply(geolocator.geocode) 
shops
```

```
shops['latitude'] = shops['location'].apply(lambda x: x.latitude)
shops['longitude'] = shops['location'].apply(lambda x: x.longitude)
```

```
shops.drop('location',axis = 'columns',inplace=True)
shops.reset_index(drop = True,inplace=True)
```

```
shops
```

```
shops.latitude.nunique()
```

```
import folium
```

```
map1 = folium.Map(
    location=[62.027287,129.732086],
    tiles='cartodbpositron',
    zoom_start=12,
)
shops.apply(lambda row:folium.CircleMarker(location=[row["latitude"], row["longitude"]]).add_to(map1), axis=1)
map1
```
