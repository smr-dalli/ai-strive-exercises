# Computer Vision Histograms

### Import statements

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Histograms

```python
# Load some images, calculate and plot their color histograms, one color at the time
# try it with different color spaces BGR, HSV, etc
bgr_image = cv2.imread('img/eclipse.jpg')
rgb_image = cv2.cvtColor(bgr_image,cv2.COLOR_BGR2RGB)
hsv_image = cv2.cvtColor(rgb_image,cv2.COLOR_RGB2HSV)
```

```python
hist_B = cv2.calcHist([bgr_image],[0],None,[255],[0,255])
plt.plot(hist_B,color = 'b');
```

```python
hist_G = cv2.calcHist([bgr_image],[1],None,[255],[0,255])
plt.plot(hist_G,color = 'g');
```

```python
hist_R = cv2.calcHist([bgr_image],[2],None,[255],[0,255])
plt.plot(hist_R,color = 'r');
```

[Hint](# "cv2.calcHist") 


```python
# Now try to plot all 3 channels at the same time, does the histogram give you an idea of what the image looks like?
colors = ['r','g','b']
for i,color in enumerate(colors):
    hist = cv2.calcHist([rgb_image],[i],None,[255],[0,255])
    plt.plot(hist,color = color)
```

```python
colors = ['r','g','b']
for i,color in enumerate(colors):
    hist = cv2.calcHist([hsv_image],[i],None,[255],[0,255])
    plt.plot(hist,color = color)
```

### Level 1

#### Image classifier

1. By using only computer vision techniques (no Neural Networks and ML are not allowed) create a binary classifier that can detect if a picutre was taken during the day, you can use the images day/night in the img folder or feel free to add your own images.

```python
image = 'img/eclipse.jpg'
rgb_image = cv2.cvtColor(cv2.imread(image),cv2.COLOR_BGR2RGB)
rgb_image.shape
```

```python
def image(img):
    rgb_image = cv2.cvtColor(cv2.imread(img),cv2.COLOR_BGR2RGB)
    hsv_image = cv2.cvtColor(rgb_image,cv2.COLOR_RGB2HSV)
    
    #standard or resize
   # resize_img = cv2.resize(rgb_image,(1100,600))
    
    ##sum of pixles
    pixles_values_sum = np.sum(hsv_image[:,:,2])
    
    ## are of pixles
    area = hsv_image.shape[0]*hsv_image.shape[1]
    
    ##classification
    if (pixles_values_sum/area) >= 127.5:
        plt.imshow(rgb_image)
        print('Picture was taken during the day')
    else:
        pass
image('img/day5.png')
    
    
```
