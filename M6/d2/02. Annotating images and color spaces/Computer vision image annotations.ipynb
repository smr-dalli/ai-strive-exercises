### To edit an image making a copy of it is always a good idea!

```python
# Load an image
image = 'img/geo.jpg'
geo = cv2.imread(image)
rgb_image = cv2.cvtColor(geo,cv2.COLOR_BGR2RGB)
# And make a copy of the original image to avoid manipulating the original image
Images = rgb_image.copy()
```

```python
plt.imshow(Images)
```

[Hint](# "Numpy .copy function") 

## Drawing

### Drawing lines

```python
# Draw some lines on top of the image 
plt.imshow(cv2.line(Images,(0,0),(600,50),(0,255,0),5));
```

[Hint](# "cv2.line function") 

### Circles

```python
# Draw some circles on top of the image
image1 = Images.copy()
plt.imshow(cv2.circle(image1,(300,150),100,(0,0,255),5));
```

[Hint](# "cv2.circle") 

### Ellipses

```python
# Draw some elipses on top of the image
plt.imshow(cv2.ellipse(image1,(300,150),(50,100),0,0,360,(255,0,0),6));
```

[Hint](# "cv2.ellipse")

### Add an overlay on top of an image

```python
# Load an image and create a copy of it you can call it overlay
# Create a rectangle with a fill color (-1) on top of the overlay image
# Combine both images setting some opacity on the overlay
# Add some text on top of the overlay and show the image on screen
```

[Hint](# "cv2.addWeighted")  

```python
image = 'img/elephant.jpg'
bgr_image = cv2.imread(image)
rgb_image = cv2.cvtColor(bgr_image,cv2.COLOR_BGR2RGB)
plt.figure(figsize = (12,12))
plt.imshow(rgb_image);
```

```python
overlay = rgb_image.copy()
cv2.rectangle(overlay,(0,0),(600,600),(0,0,0),-1)
plt.figure(figsize = (12,12))
plt.imshow(overlay)
```

```python
x = cv2.addWeighted(overlay,0.5,rgb_image,0.5,0)
cv2.putText(x,'Beautiful World',(50,50),cv2.FONT_ITALIC,2,(200,220,220),3,cv2.LINE_AA)
plt.figure(figsize = (12,12))
plt.imshow(x)
```
