```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Corners

```python
image = cv2.imread('img/chess.jpg')
plt.imshow(image);
```

```python
# We will work with the image above
# Convert it to grayscale
chess_gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)

# Detect the corners using the cornerHarris
corner = cv2.cornerHarris(chess_gray,15,7,15)


# Display the image with the detected corners 
plt.imshow(corner,cmap = 'gray');
```

```python
img = cv2.imread('img/chess.jpg')

# Convert img to grayscale
chess_gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)

# Detect the corners using goodFeaturesToTrack
corners = cv2.goodFeaturesToTrack(chess_gray,100,0.4,10)

# Draw the corners on top of img and show it on screen (make sure the corners are visible!)
for corner in corners:
    x,y = corner[0]
    x = int(x)
    y = int(y)
    cv2.rectangle(img,(x-5,y-5),(x+5,y+5),(0,255,0),-1)
plt.figure(figsize = (12,8))
plt.imshow(img)
```
