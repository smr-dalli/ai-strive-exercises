```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Template Matching

```python
# Load img/WandoScene.jpg  as the main image
img = cv2.imread('img/WandoScene.jpg')

# Load img/waldo.jpg as the template
template = cv2.imread('img/waldo.jpg') 
# Corvert the images to grayscale
gray_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
gray_template = cv2.cvtColor(template,cv2.COLOR_BGR2GRAY)

# Find the template in the image
temp = cv2.matchTemplate(gray_img,gray_template,cv2.TM_CCOEFF)
cv2.minMaxLoc(temp)

# Draw a rectagle around the the detected match
min_val,max_val,min_loc,max_loc = cv2.minMaxLoc(temp)
top_left = max_loc
bottom_right = (top_left[0] + 50,top_left[1] + 50)
cv2.rectangle(img,top_left,bottom_right,(0,255,0),6)

plt.figure(figsize = (12,8))
plt.imshow(img[:,:,::-1]);
```

```python
# Try with different waldo scenes
image = cv2.imread('img/WaldoScene3.jpg')
template = cv2.imread('img/waldo.jpg')

gray_image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
gray_template = cv2.cvtColor(template,cv2.COLOR_BGR2GRAY)

match = cv2.matchTemplate(gray_image,gray_template,cv2.TM_CCOEFF)

min_val,max_val,min_loc,max_loc = cv2.minMaxLoc(match)
top_left = max_loc
botton_right = (top_left[0] + 50,top_left[1] + 50)
cv2.rectangle(image,top_left,botton_right,(255,0,0),6)
plt.figure(figsize = (20,20))
plt.imshow(image[:,:,::-1]);
```
