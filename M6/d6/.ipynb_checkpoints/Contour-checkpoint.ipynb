```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def imshow(image):
    plt.figure(figsize = (12,8))
    plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB));
```

```python
bgr_image = cv2.imread('img/coins.jpg',cv2.IMREAD_COLOR)
gray_image = cv2.cvtColor(bgr_image,cv2.COLOR_BGR2GRAY)
_,thresh = cv2.threshold(gray_image,215,255,cv2.THRESH_BINARY_INV)

plt.imshow(thresh,cmap = 'gray')
```

```python
bgr_image_copy = bgr_image.copy()
coin_contour,hieraracy = cv2.findContours(thresh,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)
cv2.drawContours(bgr_image_copy,coin_contour,-1,(255,0,0),3)
imshow(bgr_image_copy)
```

```python
bgr_image_copy = bgr_image.copy()
sort = sorted(coin_contour,key = cv2.contourArea,reverse = True)
cv2.drawContours(bgr_image_copy,sort,3,(0,255,0),3)
imshow(bgr_image_copy)
```

```python
car = cv2.imread('img/car.jpg',cv2.IMREAD_COLOR)
gray_car = cv2.cvtColor(car,cv2.COLOR_BGR2GRAY)
plt.figure(figsize = (12,8))
plt.imshow(gray_car,cmap = 'gray');
```

```python
_,thresh = cv2.threshold(gray_car,160,255,cv2.THRESH_BINARY_INV)
plt.figure(figsize = (12,8))
plt.imshow(thresh,cmap = 'gray');
```

```python
car_copy = car.copy()
car_contour,hieraracy = cv2.findContours(thresh,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)
cv2.drawContours(car_copy,car_contour,-1,(255,0,0),3)
plt.figure(figsize = (12,8))
imshow(car_copy)
```

```python
accuracy = 0.001
car_copy1 = car.copy()
for contour in car_contour:
    perimeter = cv2.arcLength(contour,True)
    epsilon = perimeter * accuracy
    approx  = cv2.approxPolyDP(contour,epsilon,True)
    cv2.drawContours(car_copy1,[approx],-1,(0,255,0),3)
imshow(car_copy1)
```

```python
car_copy2 = car.copy()
for contour in car_contour:
    hull = cv2.convexHull(contour)
    cv2.drawContours(car_copy2,[hull],-1,(0,255,0),3)
imshow(car_copy2)
```

```python
car_copy2 = car.copy()
x,y,w,h = cv2.boundingRect(thresh)
print(x,y,w,h)
```

```python
cv2.rectangle(car_copy2,(x,y),(x+w,y+h),(0,255,0),3)
imshow(car_copy2)
```
