```
# Start writing code here...
!pip install bs4
import requests
from bs4 import BeautifulSoup
import pandas as pd

df = pd.read_csv('Best Books of the Decade_ 2000s _6922 books_.csv')
df.head(1000)
urls = df.Title_URL

def scrapper(urls):
    dic = {'Reviews':[],'Pages':[],'Year':[],'Series':[],'Awards':[],'Places':[],'Genre':[]}
    for i in range(2):
        page = requests.get(urls[i])
        soup = BeautifulSoup(page.content,'html.parser')

        review = soup.find('meta',itemprop = 'reviewCount').text.split()
        dic['Reviews'].append((review[0]))
    
        no_pages = soup.find('span',itemprop = 'numberOfPages').text.split()
        dic['Pages'].append(no_pages[0])

        published = soup.find('div',id = 'details').text.strip().split()
        year = max([int(i) for i in published if i.isnumeric()])
        dic['Year'].append(year)

        series = soup.find('h2',id= 'bookSeries').text
        if len(series)>1:
            series = True
        else:
            series = False
        dic['Series'].append(series)

        awards = soup.find('div',itemprop = 'awards').text.strip().split(',')
        dic['Awards'].append(awards)

        places = soup.find('div',id='bookDataBox').text.split('\n')
        f = False
        pl = ''
        for i in places:
            if i=='Literary Awards':
                break
            if f==True:
                if len(i)>1:
                    pl += i
            if i == 'Setting':
                f = True
        dic['Places'].append(pl)

        genre = soup.find_all('a',class_="actionLinkLite bookPageGenreLink")
        dic['Genre'].append([i.text for i in genre][:3])

    df2 = pd.DataFrame(dic)
    DF = pd.concat([df,df2],axis = 1)
    print(DF.head(2))

scrapper(urls)

```
