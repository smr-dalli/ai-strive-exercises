# Computer Vision Thresholding

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Have a look at the 2 images below
text = cv2.imread('img/t1.png')
plt.figure(figsize = (12,10))
plt.imshow(text, cmap='gray');
```

```python
text_2 = cv2.imread('img/t2.png')
plt.figure(figsize = (12,10))
plt.imshow(text_2, cmap='gray');
```

```python
# Apply threshold to the images above to make them look better, try to find the best one
text2 = text_2.copy()
text2 = cv2.cvtColor(text2,cv2.COLOR_BGR2GRAY)
_,thresh = cv2.threshold(text2,127,255,cv2.THRESH_TOZERO_INV)
plt.figure(figsize = (9,9))
plt.imshow(thresh,cmap = 'gray');
```

```python
# Apply threshold to the images above to make them look better, try to find the best one
text1 = text.copy()
text1 = cv2.cvtColor(text1,cv2.COLOR_BGR2GRAY)
_,thresh = cv2.threshold(text1,127,255,cv2.THRESH_TRUNC)
plt.imshow(thresh,cmap = 'gray');
```

```python
# Try to apply adaptiveThreshold and see if that helps, use both ADAPTIVE_THRESH_MEAN_C and ADAPTIVE_THRESH_GAUSSIAN_C
text3 = text.copy()
text3 = cv2.cvtColor(text3,cv2.COLOR_BGR2GRAY)
adaptiveThreshold = cv2.adaptiveThreshold(text3,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,5,0)
plt.imshow(adaptiveThreshold,cmap = 'gray')
```

```python
text3 = text.copy()
text3 = cv2.cvtColor(text3,cv2.COLOR_BGR2GRAY)
adaptiveThreshold = cv2.adaptiveThreshold(text3,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,5,0)
plt.imshow(adaptiveThreshold,cmap = 'gray');
```

```python
# Now apply Otsu's thresholding to the same images, always make sure to start with a copy of the original image
text3 = text.copy()
text3 = cv2.cvtColor(text3,cv2.COLOR_BGR2GRAY)
_,otsu = cv2.threshold(text3,0,255,cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
plt.imshow(otsu,cmap = 'gray');

```

```python
# Now try to first apply Gaussian Blur to the image and then Otsu thresholding
text3 = text.copy()
text3 = cv2.cvtColor(text3,cv2.COLOR_BGR2GRAY)
g_blurr = cv2.GaussianBlur(text3,(5,5),0)
retval,otsu_gb = cv2.threshold(g_blurr,0,255,cv2.THRESH_TOZERO_INV + cv2.THRESH_OTSU)
plt.imshow(otsu_gb,cmap = 'gray');
```
