```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image = cv2.imread('img/people2.jpg');
```

```python
# Given the image above, lets detect faces on the image using haar cascades
# First create a CascadeClassifier using the frontalFace classifier -> Haarcascades/haarcascade_frontalface_default.xml
face_classifier = cv2.CascadeClassifier('Haarcascades/haarcascade_frontalface_default.xml')
# Convert the image to grayscale
gray_image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)

# Using the classifiers detect all faces on the image
face_detect = face_classifier.detectMultiScale(gray_image,minNeighbors = 20)

# If no faces detected print("No faces found")
if face_detect is ():
    print('No Faces found')

# Otherwise draw a rectangle on each face that has been detected
for x,y,w,h in face_detect:
    cv2.rectangle(image,(x,y),(x+w,y+h),(255,0,0),3)
    
plt.imshow(image[:,:,::-1]);
```

```python
# Repeat the process above with your own images
img =  cv2.imread('img/faces.jpg')
gray_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

## frontal face classifier
faces_classifier = cv2.CascadeClassifier('Haarcascades/haarcascade_frontalface_default.xml')

## detect faces
faces = faces_classifier.detectMultiScale(gray_img)

if faces is ():
    print('No Face Found')
    
for x,y,w,h in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),4)

plt.figure(figsize = (12,12))
plt.imshow(img[:,:,::-1]);
```

```python
# Now do the same but this time also detect eyes in the faces using -> Haarcascades/haarcascade_eye.xml
img =  cv2.imread('img/faces.jpg')
gray_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

## frontal face classifier
faces_classifier = cv2.CascadeClassifier('Haarcascades/haarcascade_eye.xml')

## detect faces
faces = faces_classifier.detectMultiScale(gray_img)

if faces is ():
    print('No Face Found')
    
for x,y,w,h in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),4)

plt.figure(figsize = (12,12))
plt.imshow(img[:,:,::-1]);
```

```python
# Now do the same but this time also detect eyes in the faces using -> Haarcascades/haarcascade_eye.xml
img =  cv2.imread('img/faces.jpg')
gray_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

## frontal face classifier
faces_classifier = cv2.CascadeClassifier('Haarcascades/haarcascade_frontalface_default.xml')
eye_classifier = cv2.CascadeClassifier('Haarcascades/haarcascade_eye.xml')

## detect faces
faces = faces_classifier.detectMultiScale(gray_img)
eyes  = eye_classifier.detectMultiScale(gray_img)

if faces is ():
    print('No Face Found')
    
for x,y,w,h in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),4)
    
    for x,y,w,h in eyes:
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),4)
    
plt.figure(figsize = (12,12))
plt.imshow(img[:,:,::-1]);
```

```python
# Encapsulate all the logic to detect faces and eyes in a function and apply it to live video using you webcam
# Now do the same but this time also detect eyes in the faces using -> Haarcascades/haarcascade_eye.xml
#img =  cv2.imread('img/faces.jpg')

capture = cv2.VideoCapture(0)
## frontal face classifier
faces_classifier = cv2.CascadeClassifier('Haarcascades/haarcascade_frontalface_default.xml')
eye_classifier = cv2.CascadeClassifier('Haarcascades/haarcascade_eye.xml')


while (capture.isOpened()):
    _,frame = capture.read()
    if frame is not None:
        gray_img = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
        
        ## detect faces
        faces = faces_classifier.detectMultiScale(gray_img)
        eyes  = eye_classifier.detectMultiScale(gray_img)
        
        for x,y,w,h in faces:
            cv2.rectangle(frame,(x,y),(x+w,y+h),(0,0,255),4)
        
        for x,y,w,h in eyes:
            cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),4)
        
        cv2.imshow('mywindow',frame);
        
        if cv2.waitKey(30) & 0xFF == ord('q'):
            break
    else:
        print('No Frame Found')

capture.release()
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```
