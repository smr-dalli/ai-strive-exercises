```python
# importing libraries

import torch
import torchvision
from torch.utils.data import Dataset, DataLoader
from torchvision import datasets, transforms, models
import matplotlib.pyplot as plt
```

```python
preprocess = transforms.Compose([
    transforms.Resize((224,224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])
```

```python
train_data_dir = '../../dataset/train/'



trainset = datasets.ImageFolder(train_data_dir, transform=preprocess)
trainloader = torch.utils.data.DataLoader(trainset, batch_size=16, shuffle=True)
trainset.class_to_idx
```

```python
examples = iter(trainloader)
samples, labels = examples.next()
print(samples.shape, labels)
```

```python
for i in range(6):
  plt.subplot(2, 3, i+1)
  plt.imshow(samples[i][0], cmap='gray')
plt.show
```

```python
test_data_dir = '../../dataset/test/'



testset = datasets.ImageFolder(test_data_dir, transform=preprocess)
testloader = torch.utils.data.DataLoader(testset, batch_size=16, shuffle=True)
testset.class_to_idx
```

```python
examples = iter(testloader)
samples, labels = examples.next()
print(samples.shape, labels)
```

```python
for i in range(6):
  plt.subplot(2, 3, i+1)
  plt.imshow(samples[i][0], cmap='gray')
plt.show
```
