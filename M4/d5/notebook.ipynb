```
class CustomRegression(LinearRegression):
    def fit(self, X, y):
        if X.shape[1] <= 1:
            X = np.array(X)
            X = X.reshape(10,10)
        return super().fit(X, y)
    def predict(self, X):
        if X.shape[1] <= 1:
            X = np.array(X)
            X = X.reshape(10,10)
        return super().predict(X)
        
```

```
class Pets:
    def __init__(self, dogs, cats):
        self.dogs = dogs
        self.cats = cats
    def all_the_animals_have_fun(self):
        for dog, cat in zip(self.dogs, self.cats):
            dog.bark()
            cat.meow()
```

```
class Dog:
    def __init__(self, name):
        self.name = name
    def bark(self):
        print(f"{self.name} is barking")

class Cat:
    def __init__(self, name):
        self.name = name
    def meow(self):
        print(f"{self.name} is meowing")
```

```
bob = Dog("Bob")
dog = Dog("Dog")
cat = Cat("Cat")
cat1 = Cat("Cat1")
```

```
pets = Pets(dogs=[bob, dog], cats=[cat, cat1])
```

```
pets.all_the_animals_have_fun()
```
