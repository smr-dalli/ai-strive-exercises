### Exercise 1: Read the Total profit of all months and show it using a line plot.
Total profit data provided for each month. The enerated line plot must include the following properties:<br>
- X label name = Month Number
- Y label name = Total profit

The line plot graph should look like this:


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
url = 'https://pynative.com/wp-content/uploads/2019/01/company_sales_data.csv'

data = pd.read_csv(url)
data
```

```python
# Write your code here
url = 'https://pynative.com/wp-content/uploads/2019/01/company_sales_data.csv'

data = pd.read_csv(url)

overall_profit = data['total_profit']
total_months = data['month_number']
plt.plot(total_months,overall_profit)

plt.xlabel('Month Number')
plt.ylabel('Total profit')
plt.xticks(total_months)
plt.title('Total profit of all months')


```

### Exercise 2: Get the Total profit of all months and show a line plot with the following Style properties:
- Line Style dotted and Line color should be red
- Show legend at the lower right location.
- X label name = Month Number
- Y label name = Sold units number
- Add a circle marker.
- Line marker color as read
- Line width should be 3

The line plot graph should look like this:


```python
# Write your code here
fig = plt.plot(total_months,overall_profit,label = 'Total Profits of all months',color = 'r', marker = 'o', linestyle = '--', linewidth = 3 )
plt.xlabel('Month Number')
plt.ylabel('Sold units number')
plt.xticks(total_months)
plt.legend(loc = 'lower right')
plt.title('Total profit of all months')
```

```python
# Write your code here
fig, ax = plt.subplots(figsize=(10,8))

facecream_data = data['facecream']
facewash_data = data['facewash']
toothpaste_data = data['toothpaste']
bathingsoap_data = data['bathingsoap']
shampoo_data = data['shampoo']
moisturizer_data = data['moisturizer']
total_months = data['month_number']

plt.plot(total_months, facecream_data, label = 'Facecream sales',marker = 'o', linewidth =3 )
plt.plot(total_months, facewash_data, label = 'Facewash sales',marker = 'o', linewidth =3 )
plt.plot(total_months, toothpaste_data, label = 'Toothpaste sales',marker = 'o', linewidth =3 )
plt.plot(total_months, bathingsoap_data, label = 'Bathingsoap sales',marker = 'o', linewidth =3 )
plt.plot(total_months, shampoo_data,     label = 'Shampoo sales',marker = 'o', linewidth =3 )
plt.plot(total_months, moisturizer_data, label = 'Moisturizer sales',marker = 'o', linewidth =3 )

plt.xlabel('month_number')
plt.ylabel('Sold units number')
plt.xticks(total_months)
plt.title('All product sales data')
plt.legend(loc = 'upper right')

```

### Exercise 4: Read the toothpaste sales data of each month and show it using a scatter plot.

```python
# Write your code here
toothpaste_sales_data = data['toothpaste']
total_months = data['month_number']
plt.xlabel('Month Number')
plt.ylabel('Toothpaste data')
plt.xticks(total_months)
plt.scatter(total_months,toothpaste_sales_data)
plt.title('The toothpaste sales data of each month')
plt.grid(True,linestyle = '-')

```

###Â Exercise 5: Read the face cream and facewash product sales data and show it using the bar chart.
The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

The bar chart should look like this:


```python
# Write your code here
#matplotlib.pyplot.bar(x, height, width=0.8, bottom=None, *, align='center', data=None, **kwargs)
facecream_data = data['facecream']
facewash_data = data['facewash']
total_months = data['month_number']
plt.bar([w-0.25 for w in total_months],facecream_data, width = 0.25, align = 'edge', label = 'Faceceam sales')
plt.bar([w+0.25 for w in total_months], facewash_data, width = -0.25, align = 'edge', label = 'Facewash sales')
plt.xlabel('Month Number')
plt.xticks(total_months)
plt.ylabel('Facecream and Facewash Data')
plt.title('The face cream and facewash product sales data')
plt.legend(loc = 'upper left')

```

### Exercise 7: Read the total profit of each month and show it using the histogram to see the most common profit ranges.

### Exercise 8: Calculate the total sale data for last year for each product and show it using a Pie chart.
Note: In the Pie chart you should display the Number of units sold per year for each product in percentage.

The Pie chart should look like this:


```python
# Write your code here
#matplotlib.pyplot.pie(x, explode=None, labels=None, colors=None, autopct=None, pctdistance=0.6, shadow=False, labeldistance=1.1, startangle=0, radius=1, counterclock=True, wedgeprops=None, textprops=None, center=0, 0, frame=False, rotatelabels=False, *, normalize=None, data=None)
salesdata   = [data ['facecream'].sum(), data['facewash'].sum(), data ['toothpaste'].sum(), data ['bathingsoap'].sum(), data ['shampoo'].sum(), data ['moisturizer'].sum()]
plt.pie(salesdata, labels= ['facecream','facewash', 'toothpaste', 'bathingsoap', 'shampoo','moisturizer'])
plt.legend(loc = 'lower right')
plt.title('The total sale data for last year for each product ')
```
