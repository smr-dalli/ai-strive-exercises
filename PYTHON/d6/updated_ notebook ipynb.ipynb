```
# Start writing code here...
!pip install bs4
import requests
from bs4 import BeautifulSoup
import pandas as pd

page =requests.get("https://weather.com/weather/tenday/l/San+Francisco+CA?canonicalCityId=dfdaba8cbe3a4d12a8796e1f7b1ccc7174b4b0a2d5ddb1c8566ae9f154fa638c")
# Print the status code of the response
soup =BeautifulSoup(page.content, 'html.parser')

html = list(soup.children)[1]
body = list(html.children)[1]

weather = {}

day_list = []
date_list = []
high_temp_list = []
low_temp_list = []
weather_condition = []
precipation = []
wind_speed = []
weather_report = []

#data = body.find_all('details', class_="Disclosure--themeList--uBa5q",index = '0')

summary = body.find_all('p',class_ = 'DailyContent--narrative--3AcXd')
for j in range(1,11):
    description = list(summary)[j].text
    weather_report.append(description)

for i in range(1,11):
    data = body.find('div', id = f'detailIndex{i}')

    first_value = list(data.children)[0].get_text() + '.12.2020'
    day_and_date = first_value.split()

    day_list.append(day_and_date[0])
    date_list.append(day_and_date[1])

    second_value = list(data.children)[1].get_text()

    high_and_low_temp = second_value.replace('/',' ').replace('°','')
    temperature = high_and_low_temp.split()

    high_temp = temperature[0] 
    high_temperature = int((int(high_temp)-32)*(5/9))
    high_temp_list.append(str(high_temperature) + '°C')

    low_temp = temperature[1]
    low_temperature = int((int(low_temp)-32)*(5/9))
    low_temp_list.append(str(low_temperature) + '°C')

    third_value = list(data.children)[2].get_text()
    weather_condition.append(third_value)

    fourth_value = list(data.children)[3].get_text()
    precipation.append(fourth_value)

    fifth_value = list(data.children)[4].get_text()
    wind_speed.append(fifth_value)


weather['Day_name'] = day_list
weather['Date'] = date_list
weather['Max Temperature'] = high_temp_list
weather['Min Temperature'] = low_temp_list
weather['Speed of wind']= wind_speed
weather['Weather alert'] = weather_report

Dataset = pd.DataFrame(weather)
Dataset





```
