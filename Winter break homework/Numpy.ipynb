```
# Start writing code here...
#Exercise 1: Create a 4X2 integer array and Prints its attributes
import numpy as np
arr = np.empty((4,2), dtype = np.uint16)
print(arr)
print('Array shape is: ',arr.shape)
print('Array dimension: ',arr.ndim)
print('Length of the each element in bytes: ',arr.itemsize)
```

```
#Exercise 5: Add the following two NumPy arrays and Modify a result array by calculating the square of each element
arrayOne = np.array([[5, 6, 9], [21 ,18, 27]])
arrayTwo = np.array([[15 ,33, 24], [4 ,7, 1]])
new_array = np.add(arrayOne,arrayTwo)
print('addition of two arrays is:\n',new_array)
square_array = np.square(new_array)
print('Result array after calculating the square root of all elements:\n',square_array)
```

```
#Exercise 6: Split the array into four equal-sized sub-arrays
array = np.arange(10,34)
new_array = array.reshape(8,3)
print('Creating 8X3 array using numpy.arange:\n',new_array)
np.split(new_array,4)
```

```
#Exercise 7; Sort following NumPy array
#7.1- by the second row
#7.2-by the second column

sampleArray = np.array([[34,43,73],[82,22,12],[53,94,66]])
print('Printing Original array:\n',sampleArray)
print()
sorted_row = sampleArray[:,sampleArray[1,:].argsort()]
print('Sorting Original array by secoond row:\n',sorted_row)
print()
sorted_column  = sampleArray[sampleArray[:,1].argsort()]
print('Sorting Original array by secoond column:\n',sorted_column)
```

```
#Exercise 8: Following is the 2-D array. Print max from axis 0 and min from axis 1

sampleArray = np.array([[34,43,73],[82,22,12],[53,94,66]])
print('Printing Original array:\n',sampleArray)
print('Printing max Of Axis 0:',sampleArray.max(axis = 0))
print('Printing min Of Axis 1:',sampleArray.min(axis=1))
```

```
#Exercise 9: Following is the input NumPy array delete column two and insert following new column in its place.

sampleArray = np.array([[34,43,73],[82,22,12],[53,94,66]]) 
newColumn = np.array([[10,10,10]])

delete_column = np.delete(sampleArray,1,axis = 1)
print('Array after deleting column 2 on axis 1: \n',delete_column)

inser_column = np.insert(delete_column,1,newColumn,axis = 1)
print('Array after inserting column 2 on axis 1:\n',inser_column)
```
