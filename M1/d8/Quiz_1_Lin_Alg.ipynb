```python
def magnitude_complete(*inputs):
    x = []
    for args in inputs:
        x.append(args)
        return np.linalg.norm(x)



print(magnitude_complete(1) == 1.0)
print(magnitude_complete(1,2) >= 2.236)
print(magnitude_complete(1,2,3) >= 3.741)
print(magnitude_complete(1,2,3,4,5) >= 5.477)
print(magnitude_complete(1,2,3,4,5) >= 7.416)
```

```python

def magnitude_vectors(a, b):
    summation = np.add(a,b)
    #checks that size is equal
    mag = math.sqrt(summation[0]^2+ summation[1]^2)
    return (f"The magnitude of a+b = {mag}")
magnitude_vectors([5,1],[-4,-1])
```

```python

    
# or if you want to do it all in one batch:
def magdir_vectors(a, b):
    summation = np.add(a,b)

    #checks that size is equal
    angle = np.arctan(summation[1] / summation[0])
    return (f"The direction of a+b = {angle}")

    #returns the magnitude and direction in a nicely printed format 
    #"The magnitude of a + b is X and the direction is Y"
magdir_vectors([5,1],[-4,-1])
```

- If −1<α<1, will ∥αx⃗∥  increase, decrease or stay the same?

```python
def alpha_magnitude(alpha, b):
    magnitude = math.sqrt((alpha*alpha)*b[0]*b[0] + (alpha)*(alpha)*b[1]*b[1])
    return magnitude
#returns the resulting magnitude, if the direction has changed and what has happened to the vector
alpha_magnitude(2,[1,2])
```

```python
def alpha_magnitude(alpha,b):
    b = np.array(b)
    vector = alpha*b
    if np.sign(b[0]) == np.sign(vector[0]):
        print("No change in direction")
    else:
        print("direction changed from 0° to 180°")
alpha_magnitude(-20,[1,2])
```
