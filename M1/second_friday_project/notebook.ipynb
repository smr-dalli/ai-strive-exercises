```
import requests 
import re
from bs4 import BeautifulSoup
import pandas as pd

page =requests.get("https://www.imdb.com/search/title/?genres=thriller&companies=fox&sort=num_votes,desc")

soup2 =BeautifulSoup(page.content, 'html.parser')

films = soup2.find('div',{'class' : 'lister-list'})
```

```
titles = []
descriptions = []
releases = []
directors = []
ratings = []
durations = []
genres = []
stars = []
votes = []
budgets = []


for i in range(1, 40,2):
    child = list(films.children)[i]
    filmString = child.text.replace('\n', "")
    #. \ + * ? [ ^ ] $ ( ) { } = !  | : -  
    flist = re.split("\||  ", filmString)
    fset = [x for x in flist if x != '']
    if 'Metascore' in fset:
        fset.remove('Metascore')
    print(fset)
    if len(fset) > 1:
        t = fset[0]
        t = t[t.find('.') + 1:]
        rel = t[::-1][t[::-1].find(')') + 1:][:4][::-1]
        t = t[:t.find('(')]
        releases.append(int(rel))
        titles.append(t)
        durations.append(int(fset[1][:fset[1].find(' ')]))
        genres.append(fset[2])
        r = fset[3][:5]
        for l in r:
            if l.isalpha():
                r = r.replace(l,"")
                if r[1] != '.':
                    r = '9.0'

        ratings.append(float(r))
        descriptions.append(fset[4])
        directors.append(fset[5][9:])
        bud = fset[7]
        budgets.append(float(bud[bud.find('$') + 1:-1]))
        s = fset[6][7:]
        s = s[::-1]
        vot = s[:s.find(':')].replace(',', '')      
        votes.append(int(vot[::-1]))    
        s = s[s.find('V') + 1:]
        stars.append(s[::-1])
#Gross:$214.95M
boxoffice = ['375.2','369.3','141.5','330.4','746','180.3','619','414.4','232.3','226.8','407.7','358.4','347.2','481.8','80.7','487.3','226.9','348.3','552.6','388.2']
top_20 = {'title' : titles, 'description' : descriptions, 'release' : releases, 'director' :  directors, 'rating' : ratings, 'duration' : durations, 'votes': votes, 'genre' : genres, 'stars' : stars, 'budget($ Mil)': budgets,'box_office($ Mil)': boxoffice}
df = pd.DataFrame(top_20, index = list(range(1,21)))

df
```

```
df['box_office($ Mil)'] = df['box_office($ Mil)'].astype(float)
```

```
df['profit($ Mil)'] = df['box_office($ Mil)'] - df['budget($ Mil)']

import plotly.express as px
fig1 = px.bar(df, x = 'director', y = 'profit($ Mil)',title = 'Production House profits' )
fig1.show()
```

```
   #pandas stuff

#pandas dataframe with all
#normalized max_min (Rating/Duration (minutes))
#normalized0_1 (Rating / time to film)
import pandas as pd

#x normalized = (x â x minimum) / (x maximum â x minimum)



rvs = []
for i in range(0, 20):
        rv = int(votes[i] / ratings[i])
        rvs.append(rv)

rvs_n = [(x - min(rvs)) / (max(rvs) - min(rvs)) for x in rvs]
df['V/R'] = rvs_n
df



#SORTING rating/votes

df = df.sort_values(by="V/R", ascending=False)
df

```
